```python
import pandas as pd 

sdf = pd.read_csv('GSE56133stats.csv' , sep='\t')
sdf.sort_values(by='p')
```

```python
groups = ['WT Untreated',
 'WT Amp',
 'WT Gent',
 'WT Kan',
 'WT Nor',
 'WT H2O2',
 'Fur KO Untreated',
 'Fur KO Gent',
 'Hpx KO Untreated']
groups
```

```python
fsdf = sdf[sdf['p'] < 0.0000001]
fsdf
```

```python
import matplotlib.pyplot as plt 
x = fsdf['WT Untreated']
y = fsdf['Fur KO Gent']
plt.scatter(x, y)
plt.xlabel('WT Untreated')
plt.ylabel('Fur KO Gent')
```

```python
from sklearn.cluster import KMeans
import numpy as np 

X = np.matrix([*zip(x, y)])
print(X)
```

```python
kmeans = KMeans(n_clusters=4).fit(X)
print(kmeans.labels_)
plt.scatter(x, y, c=kmeans.labels_)
plt.xlabel('WT Untreated')
plt.ylabel('Fur KO Gent')
```

```python
fig = plt.figure()
fig.set_size_inches(w=30, h=30)
for i in range(0, len(groups)):
    for j in range(i+1, len(groups)):
        fig.add_subplot(len(groups)+1, len(groups)+1, i*len(groups)+j)
        x = fsdf[groups[j]]
        y = fsdf[groups[i]]
        kmeans = KMeans(n_clusters=4).fit(np.matrix([*zip(x, y)]))
        plt.scatter(x, y, c=kmeans.labels_)
        plt.xlabel(groups[j])
        plt.ylabel(groups[i])
```

```python
import scipy.cluster.hierarchy as hierarchy
import scipy.spatial.distance as dist

df1 = fsdf.drop('ID_REF', axis=1).drop('p', axis=1).drop('Unnamed: 0', axis=1)
df1
```

```python
X = df1.values
X
```

```python
import matplotlib.pyplot as plt 

fig =plt.figure()
fig.set_size_inches(w=30, h=5)
d = dist.pdist(X.T)
l = hierarchy.linkage(d)
labels = list(df1.columns.values)
dend = hierarchy.dendrogram(Z=l, labels=labels)
plt.show()
```

```python
import scipy.cluster.hierarchy as hierarchy
import scipy.spatial.distance as dist

fig = plt.figure(figsize=(18, 12))

dendTopAxis = fig.add_axes([0.20, 0.94, 0.8, 0.05]) # [left, bottom, width, height]
distTop = dist.pdist(X.T)
linkageTop = hierarchy.linkage(distTop)
labelsTop = list(df1.columns.values)
dendTop = hierarchy.dendrogram(Z=linkageTop, labels=labelsTop)

dendLeftAxis = fig.add_axes([0,0,0.15,0.92]) # [left, bottom, width, height]
distLeft = dist.pdist(X)
linkageLeft = hierarchy.linkage(distLeft)
labelsLeft = fsdf['ID_REF'].values
dendLeft = hierarchy.dendrogram(Z=linkageLeft, labels=labelsLeft, orientation='left')
xsorted = X[:,dendTop['leaves']]
xsorted = xsorted[dendLeft['leaves'], :]
heatmapAxis = fig.add_axes([0.2, 0, 0.8, 0.92]) # [left, bottom, width, height]
heatmapAxis.imshow(xsorted, aspect='auto', origin='lower', cmap='hot')
plt.show()
plt.close('all')
```
