```python
from jsonpath_ng.ext import parse
from geo import fuzzyReshape
import requests
from io import StringIO
import pandas as pd

upAccList = []
for a in [match.value for match in parse('$.NC_000913_3[*].db_xref[?@=~"^UniProtKB/Swiss-Prot:"]').find(n)]:
    upAccList.append(a.replace('UniProtKB/Swiss-Prot:', ''))

dflist = []
for batch in fuzzyReshape(upAccList, 100):
    print(batch)
    upSearchTerm = ''
    for a in batch:
        upSearchTerm = upSearchTerm + f'id:{a} OR '
    upurl = 'https://www.uniprot.org/uniprot/'
    params = {
        'query': upSearchTerm[:-4],
        'format': 'tab',
        'columns': 'id,genes,go-id,database(refseq)'
    }
    r = requests.get(upurl, params)
    print(r.url, r, r.text)
    csv = StringIO(r.text)
    dflist.append(pd.read_csv(csv, sep='\t'))
```

```python
dflist[0]
```

```python
uniprot = {'NC_000913_3': []}
for df in dflist:
    for i,r in df.iterrows():
        entry = {
            'id': str(r['Entry']).strip(),
            'gene_names': str(r['Gene names']).strip().split(),
            'goid': str(r['Gene ontology IDs']).strip().split(';'),
            'refseq': str(r['Cross-reference (refseq)']).strip().split(';')[:-1]
        }
        uniprot['NC_000913_3'].append(entry)
```

```python
import json

j = json.dumps(uniprot, indent=2)
f = open('uniprot.json', 'w')
f.write(j)
f.close()
```
