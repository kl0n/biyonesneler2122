```python
import pandas as pd 

smdf = pd.read_csv('nuc.4.4', sep='\s+', comment='#')
smdf
```

```python
smdict = {}
for index, row in smdf.iterrows():
    ssdict ={}
    for i,j in row.iteritems():
        ssdict[i] = j
    smdict[index] = ssdict
print(smdict)

print(smdict['R']['Y'])
```

```python
import numpy as np 

s1 = 'ATCCGATACTGGCTATA'
s2 = 'ATCTCTGGAGATCTATAG'

nRow = len(s1) + 1
nCol = len(s2) + 1

scoreMatrix = np.full([nRow, nCol], 0)
traceMatrix = np.full([nRow, nCol], '-')

from IPython.core.display import HTML,display
import pandas as pd

def prettyTable(data_array, row_labels,col_labels):
    """Show an HTML table from a 2d numpy array"""
    df = pd.DataFrame(data_array,index=row_labels,columns=col_labels)
    table_html = df.to_html()
    return HTML(table_html)

rowLabels = [label for label in '-'+s1]  # list comprehension
colLabels = [label for label in '-'+s2]

display(prettyTable(scoreMatrix, rowLabels, colLabels))
display(prettyTable(traceMatrix, rowLabels, colLabels))
```

```python
up_arrow = "\u2191"
right_arrow = "\u2192"
down_arrow = "\u2193"
left_arrow = "\u2190"
down_right_arrow = "\u2198"
up_left_arrow = "\u2196"

gapPenalty = -5

for i in range(nRow):
    for j in range(nCol):
        if i == 0 and j == 0:
            score = 0
            arrow = '-'
        elif i == 0:
            prevScore = scoreMatrix[i, j-1]
            score = prevScore + gapPenalty
            arrow = left_arrow
        elif j == 0:
            prevScore = scoreMatrix[i-1, j]
            score = prevScore + gapPenalty
            arrow = up_arrow
        else:
            leftScore = scoreMatrix[i, j-1] + gapPenalty
            upScore = scoreMatrix[i-1, j] + gapPenalty

            matchScore = smdict[s1[i-1]][s2[j-1]]
            diagonalScore = scoreMatrix[i-1, j-1] + matchScore

            score = max(leftScore, upScore, diagonalScore)

            if score == leftScore:
                arrow = left_arrow
            elif score == upScore:
                arrow = up_arrow
            elif score ==diagonalScore:
                arrow = up_left_arrow

        traceMatrix[i, j] = arrow
        scoreMatrix[i, j] = score

display(prettyTable(scoreMatrix, rowLabels, colLabels))
display(prettyTable(traceMatrix, rowLabels, colLabels))
```

```python
s1align = ''
s2align = ''
consensus = ''

i = len(s1)
j = len(s2)

arrow = traceMatrix[i, j]

while arrow != '-':
    s1char = s1[i-1]
    s2char = s2[j-1]
    if arrow == up_arrow:
        s2align = '-' + s2align
        s1align = s1char + s1align
        consensus = ' ' + consensus
        i = i - 1
    elif arrow == up_left_arrow:
        s1align = s1char + s1align
        s2align = s2char + s2align
        consensus = '|' + consensus
        i = i - 1
        j = j - 1
    elif arrow == left_arrow:
        s1align = '-' + s1align
        s2align = s2char + s2align
        consensus = ' ' + consensus
        j = j - 1
    arrow = traceMatrix[i, j]

print('{0}\n{1}\n{2}'.format(s1align, consensus, s2align))
```
