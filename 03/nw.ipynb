```python
import pandas as pd 

smdf = pd.read_csv('nuc.4.4', sep='\s+', comment='#')
smdf
```

```python
smdict = {}
for k,v in smdf.iteritems():
    ssdict = {}
    for i,j in v.iteritems():
        ssdict[i] = j
    smdict[k] = ssdict
smdict
```

```python
# Needle

nRows = len('-'+s1)
nCols = len('-'+s2)

scoreArray = np.full([nRows, nCols], 0)
traceback = np.full([nRows, nCols], '-')
print(scoreArray)
print(traceback)
```

```python
from IPython.core.display import HTML,display
import pandas as pd

def prettyTable(data_array, row_labels,col_labels):
    """Show an HTML table from a 2d numpy array"""
    df = pd.DataFrame(data_array,index=row_labels,columns=col_labels)
    table_html = df.to_html()
    return HTML(table_html)
```

```python
rowLabels = [label for label in '-'+s1]
colLabels = [label for label in '-'+s2]
display(prettyTable(scoreArray, rowLabels, colLabels))
```

```python
up_arrow = "\u2191"
right_arrow = "\u2192"
down_arrow = "\u2193"
left_arrow = "\u2190"
down_right_arrow = "\u2198"
up_left_arrow = "\u2196"

gapPenalty = -1

for row in range(nRows):
    for col in range(nCols):
        if row == 0 and col == 0:
            score = 0
            arrow = '-'
        elif row == 0:
            prevScore = scoreArray[row, col-1]
            score = prevScore + gapPenalty
            arrow = left_arrow
        elif col == 0:
            prevScore = scoreArray[row-1, col]
            score = prevScore + gapPenalty
            arrow = up_arrow
        else:
            leftScore =scoreArray[row, col-1] + gapPenalty
            upScore = scoreArray[row-1, col] + gapPenalty
            
            matchScore = smdict[s1[row-1]][s2[col-1]]
            diagonalLeftScore = scoreArray[row-1, col-1] + matchScore

            score = max(leftScore, upScore, diagonalLeftScore)

            if score == leftScore:
                arrow = left_arrow
            elif score == upScore:
                arrow = up_arrow
            elif score == diagonalLeftScore:
                arrow = up_left_arrow
        
        traceback[row, col] = arrow
        scoreArray[row, col] = score

display(prettyTable(scoreArray, rowLabels, colLabels))
display(prettyTable(traceback, rowLabels, colLabels))
```

```python
# alignments

s1align = ''
s2align = ''
consensus = ''

row = len(s1)
col = len(s2)
arrow = traceback[row, col]

while arrow != '-':
    s1char = s1[row-1]
    s2char = s2[col-1]
    if arrow == up_arrow:
        s2align = '-' + s2align
        s1align = s1char + s1align
        consensus = ' ' + consensus
        row -= 1
    elif arrow == up_left_arrow:
        s1align = s1char + s1align
        s2align = s2char + s2align
        if s1char == s2char:
            consensus = '|' + consensus
        else:
            consensus = '.' + consensus
        row -= 1
        col -= 1
    elif arrow == left_arrow:
        s1align = '-' + s1align
        s2align = s2char + s2align
        consensus = ' ' + consensus
        col -= 1
    arrow = traceback[row, col]
    print(row, col)

print('{0}\n{1}\n{2}'.format(s1align, consensus, s2align))
```
