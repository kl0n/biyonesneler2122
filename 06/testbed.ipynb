```python
import requests

base = 'https://www.uniprot.org/uniprot/'
params = {
    'query': 'id:P00561 OR id:P00547 OR id:P00934',
    'format': 'tab',
    'columns': 'id,genes,go'
}
r = requests.get(base, params=params)
r
```

```python
from io import StringIO
import pandas as pd 

csv = StringIO(r.text)
df = pd.read_csv(csv, sep='\t')
df
```
